## 1. 문제 풀기

> 코딩을 하기 앞서 문제를 손으로 풀어볼까요?

먼저 현재 시간은 `HH:MM` ex) 09:30, 23:30으로 시간과 분만 주어진다고 가정해보아요.  
그리고 앞당길 시간 또한 같은 형식으로 `HH:MM`으로 주어진다고 가정해보아요.  

그렇다면 우리가 계산한 후에 출력해야하는 시간은 얼마일가요? 현재 시간에서 앞당길 시간을 뺀 값이겠죠?  
만약 현재 시간이 `09:30`이고, 뺄 시간이 `01:30` 이었다면, 다음과 같이 계산될거에요.
<center>
    구해야할 시간 = 09:30 - 01:30
</center>

위의 경우는 철수조차도 간단하게 답이 `08:00`이라는 것을 알 수 있어요. 하지만 문제가 좀더 어려워진다면? 풀 수 있을까요?  

몇 가지 경우를 따져봐요.  

1. 시간에서 받아올림이 있는 경우
    다음의 경우를 생각해봐요

    <center>
        01:00 - 07:00
    </center>

    이 경우, 우리는 다음처럼 01:00을 25:00으로 만든 뒤, 07:00을 빼주면 18:00이라는 결과를 얻을 수 있어요.

2. 분에서 받아올림이 있는 경우
    반대로 분도 마찬가지에요.

    <center>
        01:30 - 00:40
    </center>

    이 경우, 우리는 다음처럼 01:30을 00:70으로 만든 뒤, 00:40을 빼주면 00:30이라는 결과를 얻을 수 있어요.

사실 두가지 케이스 모두 원리는 똑같아요. 바로 받아올림이죠. 초등학교때 배운 기초적인 내용이지만, 과연 이걸 코딩으로 구현한다면 어떻게 될까요?

### 2.2. 조건문

**조건문이란?**
* 조건에 따라 다르게 동작을 실행할 수 있도록 작성한 코드입니다.
* 조건문을 사용해, 조건이 참인 경우와 거짓인 경우를 나누어 각각 다른 코드가 실행되도록 명령할 수 있습니다.  


먼저 다음의 코드를 읽어볼까요?

```python
a = 23
b = 22

if a > b:
    print('a가 b보다 큽니다')
elif a > b:
    print('a가 b보다 작습니다')
else:
    print('a와 b가 같습니다.')
```


코드는 다음과 같이 여러 부분으로 쪼개볼 수 있어요.
1. 비교문
    
    `a > b`, `b < a`와 같은 구문들은 수학시간에 배웠죠? 바로 a가 b보다 크냐 작냐를 판정하는 비교문이에요. 비교문에 쓰이는 `>`, `<`와 같은 것을 비교연산자라고 해요.
    비교 연산자에는 여러 종류가 있어요.  
    **비교 연산자의 종류**
    비교 연산자의 종류 | 설명 | 예시 | True of False
    ---|:---:|---|---
    a>b | a가 b 초과일 때 참 | 17>5 | True
    a>=b | a가 b 이상일 때 참 | 2>=4 | False
    a==b | a와 b가 동일할 때 참 | 3==3 |True
    a!=b | a와 b가 다른 경우 참 | 7!=7 | False
    a<b | a가 b 미만일 때 참 | 3<4 |True
    a<=b | a가 b 이하일 때 참 | 7<=9 | True

    **(주의) a와 b가 동일함을 나타내는 연산자 == 를 변수에 값을 대입해주는 대입 연산자 = 과 헷갈리지 않게 주의해야 합니다.**  
    - a == b : a와 b가 동일할 때 참을 반환합니다.   
    - a = b : 변수 a에 값 b를 대입해줍니다. (참고 : 이 코드 이전에 b가 선언되지 않았다면 오류가 발생합니다.)  

    **(주의) >=, <=, != 에서 "="는 반드시 오른쪽에 작성되어야 합니다. (=>, =<, =!와 같이 작성하지 않도록 주의하세요.)**

2. 조건문
    
    조건을 만족하는 경우, 그 안에 있는 코드들이 실행돼요. 예를 들어 아래 예시는 `a가 b보다 큽니다`가 출력될거에요.
      
    반대로 a가 b보다 작은 경우라면, 아래와 같이 `a가 b보다 작습니다`가 출력될거에요.  
      
    만약 크지도 작지도 않은 같은 경우라면, 위와 같이 `a와 b가 같습니다`가 출력될거에요.  
      


## 3. 구현하기

### 3.1. 시, 분 입력받기

우리가 입력받아야 할 시간은 두가지 입니다. 바로 현재 시간과 뺄 시간이죠. `2.1`에서 배운 `input()` 함수를 사용해 시와 분을 입력받아볼까요?

**주의!**
input은 문자열을 반환하니 이를 연산할 수 있도록 정수로 변환해줘요. 정수 변환에는 `int()`라는 함수가 사용됩니다. 문자열을 입력받으면 그에 해당하는 정수를 반환해줘요.

```python
one_string = "1"
one = int(one_string) # one = 1
```

````python
current_hour = int(input())
current_minute = int(input())
subtract_hour = int(input())
subtract_minute = int(input())

print(current_hour, current_minute)
print(subtract_hour, subtract_minute)
````

### 3.2. 시 빼기

먼저 현재 시간에서 시를 빼줄게요. 예를 들어, 현재 시간이 `01:30`이고 빼야할 시간이 `03:30`이라면 `01:30`에서 `03:00`부터 뺀다는 의미에요.  
원리는 어렵지 않죠? 그렇다면 아래 그림처럼 시를 뺄 수 있는 코드를 작성해주세요!

힌트: 분은 시에 영향을 받지 않으므로 시만 고려하면 된다.

````python
result_hour = current_hour - subtract_hour
if (result_hour < 0):
    result_hour += 24

print(result_hour)
````

<center>

</center>
